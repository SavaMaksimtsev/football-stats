```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from pyfonts import load_font

prop = fm.FontProperties(fname='/Users/zaytsev/GoogleSans-Regular.ttf')

data = {
    'team': ['Байер', 'Бавария', 'Штутгарт', 'Боруссия Д', 'РБ Лейпциг', 
             'Айнтрахт', 'Аугсбург', 'Хоффенхайм', 'Фрайбург', 'Вердер', 
             'Хайденхайм', 'Боруссия М', 'Унион', 'Вольфсбург', 'Бохум', 
             'Майнц', 'Кёльн', 'Дармштадт'],
    'xG': [59.59, 78.81, 61.55, 55.47, 56.37, 
           38.76, 39.00, 40.87, 40.43, 33.79, 
           31.83, 40.57, 34.22, 33.22, 37.75, 
           33.21, 29.87, 28.56],
    'xGA': [24.82, 25.80, 30.14, 40.19, 31.78, 
            42.19, 44.54, 51.53, 46.79, 43.22, 
            50.88, 47.69, 41.81, 39.71, 54.12, 
            39.05, 58.46, 61.14]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.gcf().set_facecolor('#181818')
plt.gca().set_facecolor('#181818')

colors = ['#4EA075' if team == 'Вольфсбург' else '#E21F71' for team in df['team']]

plt.scatter(df['xG'], df['xGA'], color=colors)
plt.gca().invert_yaxis()
plt.xlabel('xG', fontproperties=prop, color='white')
plt.ylabel('xGA', fontproperties=prop, color='white')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_color('white')
plt.gca().spines['bottom'].set_color('white')
plt.gca().tick_params(axis='x', colors='white')
plt.gca().tick_params(axis='y', colors='white')

for i, row in df.iterrows():
    plt.text(row['xG'], row['xGA'], row['team'], fontproperties=prop, fontsize=9, ha='left', color='white')

plt.title('Соотношение xG и xGA в Бундеслиге к моменту увольнения Ковача', fontproperties=prop, fontsize=15, color='white')
plt.figtext(0.215, 0.02, 'Источник данных: Understat', ha='center', fontproperties=prop, fontsize=10, color='white')

plt.show()
```
