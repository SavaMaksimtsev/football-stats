```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
import pandas as pd

prop = fm.FontProperties(fname='//Users/zaytsev/GoogleSans-Regular.ttf')

data = {
    "Команда": [
        "Арсенал", "Брайтон", "Тоттенхэм", "Астон Вилла", "Ипсвич",
        "Ман Юнайтед", "Ман Сити", "Челси", "Кристал Пэлас", "Ньюкасл",
        "Ливерпуль", "Вулверхэмптон", "Ноттингем", "Эвертон", "Брентфорд",
        "Лестер", "Саутгемптон", "Фулхэм", "Борнмут", "Вест Хэм"
    ],
    "2023-24": [
        23, 41, 37, 39, 0, 45, 26, 43, 37, 41,
        35, 26, 36, 33, 29, 0, 0, 22, 25, 26
    ],
    "2024-25": [
        21, 20, 19, 19, 18, 17, 16, 16, 14, 14,
        12, 12, 10, 10, 10, 10, 9, 8, 7, 4
    ]
}

df = pd.DataFrame(data)

df_sorted = df.sort_values(by="2024-25", ascending=True)

teams = df_sorted["Команда"]
values_2023_24 = df_sorted["2023-24"]
values_2024_25 = df_sorted["2024-25"]

y = np.arange(len(teams))

plt.figure(figsize=(10, 8))

plt.barh(y - 0.2, values_2023_24, height=0.4, color='#67139D', label='2023-24')
plt.barh(y + 0.2, values_2024_25, height=0.4, color='#4EA075', label='2024-25')

plt.yticks(y, teams, fontproperties=prop)
plt.xlabel('', fontproperties=prop)
plt.title('Кол-во травм у клубов АПЛ в сезоне-2024/25 и 2023/24', fontproperties=prop, size=16)
plt.legend()
plt.gca().spines['top'].set_color('white')
plt.gca().spines['right'].set_color('white')

plt.tight_layout()
plt.savefig("epl_injuries.png")
plt.show()
```
